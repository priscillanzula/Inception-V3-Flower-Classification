```python
import kaggle
import os

# Authenticate (make sure kaggle.json is in the right place)
kaggle.api.authenticate()

# Create data directory
os.makedirs("data", exist_ok=True)

# Download the flowers dataset
kaggle.api.dataset_download_files('alxmamaev/flowers-recognition', path='data')

print("✅ Download completed!")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing import image
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Quick version / device check
print('TensorFlow version:', tf.__version__)
print('GPUs found:', tf.config.list_physical_devices('GPU'))
```

```python
# setting directory
# flow_from_directory reads images from folder-structured dataset. 
# We use validation_split=0.2 to keep 20% for validation.
data_dir = r"C:\Users\priscillah\documents\images\data\flowers"


# settings
batch_size = 16
img_size = (299, 299) 


# Data generators (simple preprocessing + validation split)
datagen = ImageDataGenerator(preprocessing_function=preprocess_input, validation_split=0.2)

train_generator = datagen.flow_from_directory(
data_dir,
target_size=img_size,
batch_size=batch_size,
class_mode='categorical',
subset='training',
shuffle=True
)
val_generator = datagen.flow_from_directory(
data_dir,
target_size=img_size,
batch_size=batch_size,
class_mode='categorical',
subset='validation',
shuffle=False
)
# Class names 
class_labels = list(train_generator.class_indices.keys())
print('Classes:', class_labels)
```

### InceptionV3 Model

```python
# Load InceptionV3 base (pretrained on ImageNet), excluding top layers
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(299,299,3))

# classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
outputs = Dense(train_generator.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=outputs)

# Freeze base model,start by training only the top layers
for layer in base_model.layers:
    layer.trainable = False

model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

### Model Training

```python
epochs = 5 
history = model.fit(train_generator, validation_data=val_generator, epochs=epochs)
```

### Validation accuracy & Loss Plots

```python
# Accuracy
plt.figure(figsize=(8,4))
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


# Loss
plt.figure(figsize=(8,4))
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

##### Accuracy curve
Training accuracy:Rises quickly from 0.74 to 0.92.

Validation accuracy: Starts 0.88, dips slightly around epoch 2, then recovers to 0.89.

##### Loss curve 
Train loss: Starts high (0.75) and decreases steadily to 0.25 by epoch 4.

Validation loss: Starts lower (0.45) and also decreases smoothly, staying slightly above training loss.

Thus, no overfitting as both training and validation losses are going down.

```python
### Save the Model
model.save('inception_flower_model.h5')
print('Saved model to inception_flower_model.h5')
```

### Evaluation on the validation set 

```python
# labels and predictions
y_true = val_generator.classes
y_proba = model.predict(val_generator, verbose=1)
# predicted classes
y_pred = np.argmax(y_proba, axis=1)

# Classification report
print('\nClassification Report:')
print(classification_report(y_true, y_pred, target_names=class_labels))

# Per-class accuracy (recall is the per-class true positive rate)
report = classification_report(y_true, y_pred, target_names=class_labels, output_dict=True)
print('\nPer-class accuracy (recall):')
for c in class_labels:
    print(f"{c}: {report[c]['recall']:.2f}")
```

#### Metrics Explained:
1. Precision: How many predicted flowers of a type were actually correct.

2. Recall: How many real flowers of that type were correctly identified (this is the “accuracy per class”).

3. F1-score: Balance between precision and recall.

#### Class Performance
- Daisy has a score of 0.95 (95%) recall which shows an excellent recognition. The model almost never misses a daisy.

- Dandelion has a score 0.93 (93%) recall which shows a strong performance, the model identifies dandelions well.

- Rose has a score of 0.88 (88%) recall which shows a good performance, but slightly weaker as some roses are misclassified as other flowers.

- Sunflower has a score of 0.87 (87%) recall which shows a decent, but slightly lower performance compared to daisy/dandelion. The model is likely to have confusion with daisy or othe yellow colored images.

- Tulip has a score 0.81 (81%) recall which is the lowest. The model struggles most with tulips, possibly confusing them with roses due to shape/overlapping colors.                               
                                 

#### Accuracy per class plot

```python
accuracies = [report[c]['recall'] for c in class_labels]
plt.figure(figsize=(8,4))
plt.bar(class_labels, accuracies, color ='green')
plt.ylim(0,1)
plt.title('Per-class Accuracy (Recall)')
plt.ylabel('Accuracy')
plt.xticks(rotation=45)
plt.show()
```

### Confusion Matrix

```python
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6,6))
plt.imshow(cm, interpolation='nearest')
plt.title('Confusion Matrix')
plt.colorbar()

tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels, rotation=45)
plt.yticks(tick_marks, class_labels)

# Annotate counts
thresh = cm.max() / 2
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j, i, format(cm[i, j], 'd'), ha='center', va='center',
                 color='white' if cm[i, j] > thresh else 'black')

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()
plt.show()
```

From the matrix, the model achieves high accuracy performance, with 196 dandelions, 159 tulips, 145 daisies, 137 roses, and 127 sunflowers correctly identified.
The main challenge lies in differentiating tulips and roses, which are often confused, while sunflowers also overlap with daisies. 
Overall, the model performs strongly but struggles with separating some visually similar classes.

#### Sample predictions and misclassified images

```python
# First 5 validation images with predictions
print('Sample predictions:')
for i in range(5):
    img_path = val_generator.filepaths[i]
    img = image.load_img(img_path, target_size=img_size)
    arr = image.img_to_array(img)
    x = np.expand_dims(arr, axis=0)
    x = preprocess_input(x)
    preds = model.predict(x)
    pred_label = class_labels[np.argmax(preds)]
    true_label = class_labels[y_true[i]]
    plt.imshow(image.array_to_img(arr))
    plt.title(f'True: {true_label} | Predicted: {pred_label}')
    plt.axis('off')
    plt.show()

# Find misclassified indices and plot some
mis_idx = np.where(y_pred != y_true)[0]
print('\nTotal misclassified on validation:', len(mis_idx))

num_to_show = min(12, len(mis_idx))
if num_to_show > 0:
    plt.figure(figsize=(12, 8))
for i, idx in enumerate(mis_idx[:num_to_show]):
    plt.subplot(3, 4, i+1)
    img = image.load_img(val_generator.filepaths[idx], target_size=(150,150))
    plt.imshow(img)
    plt.title(f'T:{class_labels[y_true[idx]]}\nP:{class_labels[y_pred[idx]]}', fontsize=9)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
else:
    print('No misclassified samples to show!')
```
